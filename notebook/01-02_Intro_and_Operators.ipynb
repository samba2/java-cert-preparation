Order: byte > short > char > int > long > float > double

````java
int i = 12;
long j = i;
System.out.println(j);

float x = 42;
double y = x;
System.out.println(y);
````

### Shifting

Syntax: `value <operator> <number_of_times>`

````java
int a = 0b11110000;
System.out.println(Integer.toBinaryString(a >> 2))
````

````java
int optionA = 0b001;
int optionB = 0b010;
int optionC = 0b100;

void doSomething(int bitMask) {
    if ((bitMask & optionA) != 0b000) { // binary representation of '0'
        System.out.println("Option A chosen");
    }
    if ((bitMask & optionB) != 0) {
        System.out.println("Option B chosen");
    }
    if ((bitMask & optionC) != 0) {
        System.out.println("Option C chosen");
    }
}

//    001
//    100
// OR 101   > this is the past in bitmask
//
// option a?
//    101   // bitmask
//AND 001   // optionA
//    001   > yes
//
// option b?
//    101  // bitmask
//AND 010  // optionB
//    000  > no
//
// option c?
//    101  // bitmask
//AND 100  // optionC
//    100  > yes
//
doSomething(optionA | optionC)
````

````java
doSomething(optionB)
````

### Switch Statement

In Java 11 support for:
- byte
- short
- int
- char
- **String**
- **enum**

If no `break` then it falls through.

````java
var beer = "Ale";
var taste = "";

switch (beer) {
    case "Ale":
        taste = "yummy";
        break;
    case "Pils":
        taste = "alright";
        break;
    default:
        System.out.println("unknown beer");
}

taste
````
