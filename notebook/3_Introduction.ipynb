- excessive imports like java.lang.* are removed at compile time - it does not do harm at least to the compiler
- access modifiers
	- protected - visible to 
		- class which are in the same package or 
		- to subclasses which can be in any other package !
	- <default> - visible only to classes which are in the same package
    

- 
    - -cp - classpath
	- -d - output dir

- execute
	- javac -cp ... demos.Whatever <arguments>
		- class path can contain 
			- folders
			- archives
			- single class files
	- java 11: java /some/Whatever.java

- comments
	- //
	- / * ... */
	- /** ... */ - documentation comments (aka Javadoc)

- variable name: 4abc is invalid - no leading number
