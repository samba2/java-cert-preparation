## Final Array in Classes

- `final` is only preventing re-assignment but allows alteration of content

```java
class A { 
    final long[] value = {0}; 
}

var a = new A();
++a.value[0];
Arrays.toString(a.value)
```

### sort + toString

```java
String[] alphabet = {"Zulu", "Alpha", "Yankee"};

// sort via comparable (implemented by the object)
// mutates array!
Arrays.sort(alphabet);

Arrays.toString(alphabet)
```

```java
public class LengthComparator implements Comparator<String> {
    public int compare(String s1, String s2) {
        return s1.length() - s2.length();
    }
}

String[] stuff = {"CCC", "BB", "A"};

// sort via external comperator
Arrays.sort(stuff, new LengthComparator());

Arrays.toString(stuff);
```

# Loops

3x counting to 10

```java
int i = 0;
while (i < 10) {
    // do something
    i++;
}
```

```java
// guarantees that body is executed at least once
int i = 0;
do {
    // ... something
    i++;
} while (i < 10)
```

```java
for (int i = 0; i < 10; i++) {
   // do something
}
```

```java
// to be prepared for exam question
// while loop structure as for loop
int i = 0;
for (; i < 10; ) {
    // do something
    i++;
}
```

## for vs forEach

```java
int[] numbers = {1,2,3};

for (int i = 0; i < numbers.length; i++) {
    // do something 
    i++;
}


// auto-exracts values
for (int i : numbers) {
    // do something
}
```
